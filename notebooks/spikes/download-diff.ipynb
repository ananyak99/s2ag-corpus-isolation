```python
import os

from dotenv import load_dotenv
import requests
```

```python
load_dotenv()
api_key = os.getenv('S2_API_KEY')
headers = {"x-api-key": api_key}
```

```python
headers
```

```python
start_release_id = "2025-05-06"
end_release_id = "latest"
dataset_name = "citations"
```

```python
url = f"https://api.semanticscholar.org/datasets/v1/diffs/{start_release_id}/to/{end_release_id}/{dataset_name}"
```

```python
url
```

```python
response = requests.get(url, headers=headers)
```

```python
response
```

```python
diffs = response.json()['diffs']
```

```python
diffs
```

```python
first_diff = diffs[0]
```

```python
updates = first_diff['update_files']
deletions = first_diff['delete_files']
```

```python
len(updates)
```

```python
len(deletions)
```

```python
updates[0]
```

```python
u0 = requests.get(updates[0])
```

```python
u0.status_code
```

```python
with open('u0.gz','wb') as uf:
    uf.write(u0.content)
```

```python
d0 = requests.get(deletions[0])
```

```python
with open('d0.gz','wb') as df:
    df.write(d0.content)
```
